```python
fare_survived = df.loc[df['Survived']==1, "Fare"]

fare_survived.plot.hist(fc=(0, 0.5, 1, 0.5), normed=1, label="Survived")

fare_did_not_survive = df.loc[df['Survived']==0, "Fare"]

fare_did_not_survive.plot.hist(fc=(1, 0, 1, 0.5), normed=1, label="Did not Survive")
plt.xlabel("Fare")
plt.ylabel("Percentage of Passengers")
plt.legend(loc='upper right')
plt.title("Fare Distribution of Survivors and Non-Survivors")
```

```python
survivors_1 = df.loc[(df["Survived"]==1)&(df["Pclass"]==1), "Age"]
survivors_2 = df.loc[(df["Survived"]==1)&(df["Pclass"]==2), "Age"]
survivors_3 = df.loc[(df["Survived"]==1)&(df["Pclass"]==3), "Age"]

survivors_1.plot.hist(fc=(0, 0, 1, 0.5), label="Class 1")
survivors_2.plot.hist(fc=(1, 0, 0, 0.5), label="Class 2")
survivors_3.plot.hist(fc=(0, 1, 0, 0.5), label="Class 3")
plt.xlabel("Age")
plt.legend(loc='upper right')
plt.title("Age Distribution of Survivors by Ticket Class")
```
